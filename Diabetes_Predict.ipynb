```python
# Create a copy
df_clean=df.copy()
```

```python
df_clean.head()
```

```python
#FEATURE ENGINEERING (Diabetes Risk Analysis)

#BMI Category (Medical Insight)

df_clean['BMI_Category']=pd.cut(df['BMI'], 
bins=[0, 18.5, 24.9, 29.9, 100],
labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

#Age Group

df_clean['Age_Group']=pd.cut(df['Age'], 
bins=[0, 25, 35, 45, 55, 100],
labels=['18–25', '26–35', '36–45', '46–55', '56+'])

#Glucose Risk Category (Very Important)

df_clean['Glucose_Level']=pd.cut(df['Glucose'],
bins=[0, 69, 99, 125, 300],
labels=['Low', 'Normal', 'Prediabetic', 'Diabetic'])

#Blood Pressure Category
df_clean['BP_Category'] = pd.cut(
    df_clean['BloodPressure'],
    bins=[0, 59, 79, 89, 300],
    labels=['Low', 'Normal', 'High', 'Very High'])

#Pregnancy Risk Flag
df_clean['High_Pregnancy_Risk'] = df_clean['Pregnancies'].apply(
    lambda x: 1 if x >= 5 else 0
)

#Family History Risk Flag (DPF)
df_clean['Family_History_Risk'] = df_clean['DiabetesPedigreeFunction'].apply(
    lambda x: 1 if x >= 1 else 0
)

#Obesity + High Glucose (Interaction Feature)
df_clean['Obese_High_Glucose'] = (
    (df_clean['BMI'] >= 30) & (df_clean['Glucose'] >= 126)
).astype(int)

#Total Risk Score
df_clean['Risk_Score'] = (
    (df_clean['Glucose'] >= 126).astype(int) +
    (df_clean['BMI'] >= 30).astype(int) +
    (df_clean['Age'] >= 45).astype(int) +
    (df_clean['BloodPressure'] >= 90).astype(int) +
    (df_clean['Family_History_Risk'])
)

#Final Feature Preview
df_clean.head()
```

```python
#EXPLORATORY DATA ANALYSIS (EDA)
#Target Variable Distribution (Outcome) 
import matplotlib.pyplot as plt

df_clean['Outcome'].value_counts().plot(kind='bar')
plt.title('Diabetes Outcome Distribution')
plt.xlabel('Outcome (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
#Age Distribution vs Outcome
df_clean.groupby('Age_Group')['Outcome'].mean().plot(kind='bar')
plt.title('Diabetes Rate by Age Group')
plt.ylabel('Diabetes Probability')
plt.show()
```
